```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2.1 2nd derivative with respect to x

```python
def f(x,s=1):
    y = 1/(np.sqrt(2*np.pi)*s**3)*(x**2/s**2-1)*np.exp(-x**2/2/s**2)
    return y
```

```python
N = 6
x = np.arange(-N,N,0.001)
y = f(x)
```

```python
plt.figure()
plt.plot(x,y)
plt.title("2nd derivative with respect to x ")
plt.grid()
plt.xlabel("x")
plt.xlim((-N,N))
plt.ylabel("g\"(x)")
plt.ylim((-0.4,0.2))
```

# 2.2 plot the difference of Gaussians in 1-D

```python
def g(x,s):
    y = 1/(np.sqrt(2*np.pi)*s)*np.exp(-x**2/2/s**2)
    return y
```

```python
def D(x,s,k):
    y = (g(x,k*s)-g(x,s))/(k*s-s)
    return y
```

```python
N = 5
for k in np.arange(1.2,2.1,0.2):
    x = np.arange(-N,N,0.001)
    y = D(x,1,k)
    plt.figure()
    plt.plot(x,y)
    plt.title("the difference of Gaussians in 1-D given k={:.1f}".format(k))
    plt.grid()
    plt.xlabel("x")
    plt.xlim((-N,N))
    plt.ylabel("D(x,1,{:.1f})".format(k))
    plt.ylim((-0.4,0.2))
```

```python
k = 1.001
x = np.arange(-N,N,0.001)
y = D(x,1,k)
plt.figure()
plt.plot(x,y)
plt.title("the difference of Gaussians in 1-D given k={:.3f}".format(k))
plt.grid()
plt.xlabel("x")
plt.xlim((-N,N))
plt.ylabel("D(x,1,{:.3f})".format(k))
plt.ylim((-0.4,0.2))
```
